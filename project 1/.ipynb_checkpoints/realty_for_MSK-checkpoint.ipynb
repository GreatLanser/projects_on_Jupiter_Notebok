```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
"""Прогноз, какая точка будет наиболее выгодна для открытия. Входящие данные: 
данные по сдаваемым помещениям в мск (файл 'realty_data_for_MSK.csv'). 
В файле умеются следующие данные по точке:"""
#  floor - этаж;
#  area_unit - единицы измерения площади. Везде стоит SQUARE_METER;
#  total_floors - общее число этажей;
#  area - площадь поещения;
#  price - цена аренды в месяц, в руб.;
#  offer_type - тип объявления. Здесь везде стоит RENT, то есть все помещения сдаются;
#  address - фактический адрес нахождения точки;
#  commercial_type - тип сдаваемого помещения. Может быть под офис (OFFICE), свободного назначения(FREE_PURPOSE), \
#      склады (WAREHOUSE) и точки продажи (RETAIL)
#  longitude - долгота
#  build_year - год постройки здания
#  latitude - широта
#  distance - расстояние от центра
#  traffic - число прохожих за час
#  already_taken - сдано или нет (0 - занято, 1 - свободно)
#  competitors - число конкурентов
```

```python
file = 'realty_data_for_MSK.csv'
df = pd.read_csv(file)
```

```python
sns.scatterplot(x=df['longitude'], y=df['latitude'], hue=df['distance'], s=10)
```

```python
#  минимальная площадь
min_required_area = 20
#  максимальная дистанция от центра
max_required_distance = 10
#  максимальная цена помещения
max_affordable_price = 100000
```

```python
condition = (
    (df['floor'] == 1) &
    (df['area'] >= min_required_area) &
    (df['distance'] <= max_required_distance) &
    (df['competitors'] <= 1) &
    (df['offer_type'] == "RENT") &
    (df['price'] <= max_affordable_price) &
    df['commercial_type'].isin(['FREE_PURPOSE', 'RETAIL'])
)
df_with_conditions = df[condition]
df.shape, df_with_conditions.shape
```

```python
sns.scatterplot(x=df_with_conditions['longitude'], 
                y=df_with_conditions['latitude'], 
                hue=df_with_conditions['distance'], s=10)
```

```python
#  сколько часов в день открыты
open_hours_number = 12
#  среднее отношение посетителей к прохожим
traffic2visitor_average_ratio = 1/50
#  пессимистическое отношение посетителей к прохожим
traffic2visitor_pessimistic_ratio = 1/50
#  среднее отношение посетителей к покупателям
visitor2customers_average_ratio = 1/5
#  пессимистическое отношение
visitor2customers_pessimistic_ratio = 1/20
#  средняя сумма заказа
average_order_value = 1000
#  пессимистическая сумма заказа
pessimistic_average_order_value = 500
#  наценка на товар
trade_margin = 0.33
#  сколько дней в месяц работают
days_in_month = 30

# множитель для прогноза средней выручки
income_multiplier_average = (trade_margin *
                             average_order_value *
                             visitor2customers_average_ratio *
                             traffic2visitor_average_ratio *
                             open_hours_number *
                             days_in_month)

# множитель для прогноза пессимистической выручки
income_multiplier_pessimistic = (trade_margin *
                                 pessimistic_average_order_value *
                                 traffic2visitor_pessimistic_ratio *
                                 visitor2customers_pessimistic_ratio *
                                 open_hours_number *
                                 days_in_month)
```

```python
#  число работников
number_of_employees = 2
#  зарплата каждого работника + налоги в 43%
salary = 50 * 1.43

#  множитель для расчета расходов
addition_to_expenses = salary * number_of_employees
```

```python
#  средняя выручка с точки
df_with_conditions['average_income'] = df_with_conditions['traffic'] * income_multiplier_average
#  пессимистическая выручка с точки
df_with_conditions['pessimistic_income'] = df_with_conditions['traffic'] * income_multiplier_pessimistic
#  убытки
df_with_conditions['expenses'] = df_with_conditions['price'] + addition_to_expenses
# средняя прибыль с точки в месяц
df_with_conditions['average_profit'] = df_with_conditions['average_income'] - df_with_conditions['expenses']
df_with_conditions['pessimistic_profit'] = df_with_conditions['pessimistic_income'] - df_with_conditions['expenses']
```

```python
#  минимальная ожидаемая прибыль
minimal_expected_profit = 200000

#  все точки с прибылью, большей ожидаемой
df_with_positive_profit = df_with_conditions[df_with_conditions['average_profit'] > minimal_expected_profit]
```

```python
# вывод адресов точек, средняя и пессимистическая прибыль
for address, average_profit, pessimistic_profit in zip(df_with_positive_profit.address,
                                                       df_with_positive_profit.average_profit,
                                                       df_with_positive_profit.pessimistic_profit):
    print(address, '\n', average_profit, '\n', pessimistic_profit,'\n', '-' * 20, '\n')
df_with_positive_profit.shape
```

```python
# отсеивание точек с отрицательной пессимистичной прибылью. Осталось 2 варианта!
df_with_positive_profit = df_with_positive_profit[df_with_positive_profit['pessimistic_profit'] > 0]
for address, average_profit, pessimistic_profit in zip(df_with_positive_profit.address,
                                                       df_with_positive_profit.average_profit,
                                                       df_with_positive_profit.pessimistic_profit):
    print(address, '\n', average_profit, '\n', pessimistic_profit,'\n', '-' * 20, '\n')
```
